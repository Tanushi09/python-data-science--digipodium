```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
y = [21,345,56,12,5,47,1]
plt.plot(y)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.pie(y)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1,ncols=1)
plt.show()
```

```python
t = np.arange(-3*np.pi,3*np.pi,0.2)
x = np.sin(t)
y = np.cos(t)
fig,ax= plt.subplots(nrows=1,ncols=2,figsize=(15,2))
ax[0].plot(t,x,'r')
ax[1].bar(t,y)
plt.show()
```

```python
x = [100,200,300,400,500,500,700,800,900,1000]
y = [23,45,1,34,65,2,2,46,34,65]
plt.plot(x,y,linewidth=3.0,color='pink',marker='p',markersize='10',markerfacecolor='blue',markeredgecolor='pink',linestyle=':')
```

```python
plt.plot(x,y,linewidth=3.0,color='pink',marker='p',markersize='10',markerfacecolor='blue',markeredgecolor='pink',linestyle=':')
plt.title("some graph showing something",fontdict={'fontsize':20,'fontweight':'light','color':'pink'})
plt.xlabel("x axis ki values")
plt.ylabel("y axis ki values",fontdict={'color':'red'})
plt.text(800,60,"this is serious")
plt.annotate("peak value",xy=(500,65),xytext=(600,50),arrowprops=dict(facecolor='black'))
plt.show()
```
