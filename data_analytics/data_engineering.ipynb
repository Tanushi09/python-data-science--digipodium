```python
df.isnull().sum()   #it counts the number of missing data in each column
```

```python
df['normalized-losses'].astype('float')       #to convert it into float
```

```python
df['normalized-losses'].astype('float').sum()
```

```python
mn1=df['normalized-losses'].astype('float').mean()
```

```python
df['normalized-losses'].replace(np.nan,mn1,inplace=True)
df.head()
```

```python
df['bore'].replace(np.nan,df['bore'].astype('float').mean(),inplace=True)
df['horsepower'].replace(np.nan,df['horsepower'].astype('float').mean(),inplace=True)
df['peak-rpm'].replace(np.nan,df['peak-rpm'].astype('float').mean(),inplace=True)
df['price'].replace(np.nan,df['price'].astype('float').mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['stroke'].replace(np.nan,df['stroke'].astype('float').mean(),inplace=True)
df['num-of-doors'].replace(np.nan,'four',inplace=True)
```
